```python
def get_text(text, obj='name'):
    text = literal_eval(text)
    if len(text) == 1:
        for i in text:
            return i[obj]
    else:
        s = []
        for i in text: 
            s.append(str(i[obj]))
        return ', '.join(s)
```

```python
data['name_crew'] = data['crew'].apply(get_text ,obj = "name")
data['department_crew'] = data['crew'].apply(get_text ,obj = "department")
data['gender_crew'] = data['crew'].apply(get_text ,obj = "gender")
data['job_crew'] = data['crew'].apply(get_text ,obj = "job")
data['profile_path_crew'] = data['crew'].apply(get_text ,obj = "profile_path")
data['id_crew'] = data['crew'].apply(get_text ,obj = "id")
```

```python
data['name_cast'] = data['cast'].apply(get_text ,obj = "name")
data['order_cast'] = data['cast'].apply(get_text ,obj = "order")
data['gender_cast'] = data['cast'].apply(get_text ,obj = "gender")
data['credit_id_cast'] = data['cast'].apply(get_text ,obj = "credit_id")
data['profile_path_cast'] = data['cast'].apply(get_text ,obj = "profile_path")
data['id_cast'] = data['cast'].apply(get_text ,obj = "id")
data['character_cast'] = data['cast'].apply(get_text ,obj = "character")
```

```python
data =pd.read_csv("data/movies_metadata.csv")
data
```

```python
def is_float(string):
    try:
        float(string)
        return True
    except ValueError:
        return False
```

```python
for i in range(data.shape[0]):
    if is_float(data.loc[i, 'production_countries']) or is_float(data.loc[i, 'production_companies']):
        data.drop(index=i, inplace=True)
data.shape
```

```python
data.dropna(subset=['production_companies', 'production_countries'], inplace=True)
data.shape
```

```python
data.columns
```

```python
def json_to_arr(cell, wanted="name"):
    cell = literal_eval(cell)
    if cell == [] or (isinstance(cell, float) and cell.isna()):
        return np.nan
    result = []
    counter = 0
    for element in cell:
        if counter < 3:
            result.append(element[wanted])
            counter += 1
        else:
            break
    return result[:3]
```

```python
data.shape
```

```python

data[['genres']] = data[['genres']].applymap(json_to_arr)
data[['production_countries']] = data[['production_countries']].applymap(lambda row: 
                                                                      json_to_arr(row, "iso_3166_1"))
data[['production_companies']] = data[['production_companies']].applymap(json_to_arr)
```

```python
data = data.fillna("NANE")
```

```python
def get_new_text(text, obj='name'):
    if(text == 'NANE'):  
         return np.nan
    elif (isinstance(literal_eval(str(text)), float)):
        return np.nan
    else:
        text = literal_eval(str(text))
        for i in text:
            if(i == obj):     
                return (text[i])
```

```python
def textGeter(text, obj='name'):
    if(text == 'NANE'):  
         return np.nan
    elif (isinstance(literal_eval(str(text)), float)):
        return np.nan
    elif(isinstance(literal_eval(str(text)), list)):
        s = []
        for i in text: 
            s.append(str(i[obj]))
        return ', '.join(s)
    else:
        text = literal_eval(str(text))
        for i in text:
            if(i == obj):     
                return (text[i])
```

```python
data['name_belongs_to_collection'] = data['belongs_to_collection'].apply(get_new_text ,obj = "name")
data['id_belongs_to_collection'] = data['belongs_to_collection'].apply(get_new_text ,obj = "id")
data['poster_path_belongs_to_collection'] = data['belongs_to_collection'].apply(get_new_text ,obj = "poster_path")
data['backdrop_path_belongs_to_collection'] = data['belongs_to_collection'].apply(get_new_text ,obj = "backdrop_path")
```
