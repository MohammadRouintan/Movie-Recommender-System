```python
def get_text(text, obj='name'):
    text = literal_eval(text)
    if len(text) == 1:
        for i in text:
            return i[obj]
    else:
        s = []
        for i in text: 
            s.append(str(i[obj]))
        return ', '.join(s)
```

```python
data['name'] = data['crew'].apply(get_text ,obj = "name")
data['department'] = data['crew'].apply(get_text ,obj = "department")
data['gender'] = data['crew'].apply(get_text ,obj = "gender")
data['job'] = data['crew'].apply(get_text ,obj = "job")
data['profile_path'] = data['crew'].apply(get_text ,obj = "profile_path")
data['id'] = data['crew'].apply(get_text ,obj = "id")
```

```python
data['name'] = data['cast'].apply(get_text ,obj = "name")
data['order'] = data['cast'].apply(get_text ,obj = "order")
data['gender'] = data['cast'].apply(get_text ,obj = "gender")
data['credit_id'] = data['cast'].apply(get_text ,obj = "credit_id")
data['profile_path'] = data['cast'].apply(get_text ,obj = "profile_path")
data['id'] = data['cast'].apply(get_text ,obj = "id")
data['character'] = data['cast'].apply(get_text ,obj = "character")
```

```python
newData = data.drop(["crew" ,"cast"] ,axis=1)
```

```python
newData.to_csv('newCredits.csv', index=False)
```

```python
def get_new_text(text, obj='name'):
    if(text == 'NANE'):  
         return np.nan
    elif (isinstance(literal_eval(str(text)), float)):
        return np.nan
    else:
        text = literal_eval(str(text))
        for i in text:
            if(i == obj):     
                return (text[i])
```

```python
def textGeter(text, obj='name'):
    if(text == 'NANE'):  
         return np.nan
    elif (isinstance(literal_eval(str(text)), float)):
        return np.nan
    elif(isinstance(literal_eval(str(text)), list)):
        s = []
        for i in text: 
            s.append(str(i[obj]))
        return ', '.join(s)
    else:
        text = literal_eval(str(text))
        for i in text:
            if(i == obj):     
                return (text[i])
```
