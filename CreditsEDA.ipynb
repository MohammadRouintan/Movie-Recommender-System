```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ast import literal_eval
import seaborn as sns
```

```python
data = pd.read_csv("data/NewCredits.csv")
data.head(10)
```

### Remove all null data

```python
data = data.dropna()
data
```

```python
def castColumn(columns):
    columns =  literal_eval(columns)
    columns = list(columns)
    return columns
```

```python
def memberNumber(column):
    try:
        column =  castColumn(column)
        column.sort()
    except:
        column =  stringToList(column)
        column.sort()
    returnDict = dict()
    size = 1
    temp = column[0]
    for i in range(1, len(column) ,1):
        if temp == column[i]:
            size +=1
        else :
            returnDict[column[i]]  = size
            size = 1
        temp  = column[i]
        
    return returnDict
```

```python
def toString(column):
    temp = []
    for i in range(len(data[column])):
        if i in data["id_crew"]:
            temp.append(str(data[column][i]))

    return "".join(temp)
```

```python
def stringToList(column):
    finalList = list()
    for i in range(len(data[column])):
        if i in data.index:
            for j in data[column][i].split(','):
                finalList.append(j)
            
    return finalList
```

```python
def sortBarPlot(column ,reverse = True):
    sortcolmnTop50 = dict(sorted(memberNumber(column).items(), key=lambda x:x[1], reverse=reverse)[:50])
    plt.figure(figsize=(15, 10))
    sns.barplot(x=list(sortcolmnTop50.keys()), y=list(sortcolmnTop50.values()))
    plt.xticks(rotation=90)
    plt.show()
```
