## Keywords Dataset EDA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
keywords = pd.read_csv('data/NewKeywords.csv')
keywords.head()
```

```python
# find unique key word
def is_float(string):
    try:
        float(string)
        return True
    except ValueError:
        return False

keywords_set = set()
def find_unique(keywords_str):
    if not is_float(keywords_str):
        string_list = keywords_str.split(', ')
        for string in string_list:
            keywords_set.add(string)
```

```python
keywords['name_keywords'].apply(find_unique)
```

```python
print('Number Of Unique keywords:', len(keywords_set))
```

```python
keywords_dict = {}

for keyword in keywords_set:
    keywords_dict[keyword] = 0

print(keywords_dict)
```

```python
def count(keywords_str):
    if not is_float(keywords_str):
        string_list = keywords_str.split(', ')
        for string in string_list:
            keywords_dict[string] += 1
```

```python
keywords['name_keywords'].apply(count)
```

```python
keywords_dict.values()
```

```python
max_key = max(keywords_dict, key=lambda k: keywords_dict[k])
print('The most used keyword', '\"', max_key, '\"', ':',keywords_dict.get(max_key))
```

```python
min_key = min(keywords_dict, key=lambda k: keywords_dict[k])
print('The most used keyword', '\"',min_key, '\"', ':',keywords_dict.get(min_key))
```

```python
top_50 = pd.DataFrame(sorted(keywords_dict.items(), key=lambda item: item[1], reverse=True)[:50], columns=['Word', 'Number'])
print('Top 50 in the most used keyword', top_50, sep='\n')
```

```python
bottom_50 = pd.DataFrame(sorted(keywords_dict.items(), key=lambda item: item[1])[:50], columns=['Word', 'Number'])
print('Bottom 50 in the most used keyword', bottom_50, sep='\n')
```

```python
plt.figure(figsize=(15, 10))
sns.barplot(x=top_50['Word'], y=top_50['Number'])
plt.xticks(rotation=90)
plt.show()
```
